#### using .groupby() to count  buildings per LandUse code 

```python
# manhattan
mngrouped = mn.groupby(['LandUse']).size().reset_index(name = 'counts')
mngrouped
```

```python
# brooklyn
bkgrouped = bk.groupby(['LandUse']).size().reset_index(name = 'counts')
bkgrouped
```

#### converting the landuse to integer using .astype()

```python
mngrouped['LandUse'] = mngrouped['LandUse'].astype(int)
bkgrouped['LandUse'] = bkgrouped['LandUse'].astype(int)
mngrouped.head()
```

#### merging the dataframes on land use 

```python
both = mngrouped.merge(bkgrouped, on = 'LandUse', suffixes = ('_m', '_b'))
both.head()
```

#### renaming the columns with descriptive names

```python
both.rename(columns = {'counts_m':'mn', 'counts_b':'bk'}, inplace = True)
both
```

#### replacing landuse category labels with verbose description

```python
verbose = {1:'One & Two Family Buildings', 2:'Multi-Family Walk-Up Buildings', 3:'Multi-Family Elevator Buildings',
            4:'Mixed Residential & Commercial Buildings', 5:'Commercial & Office Buildings', 6:'Industrial & Manufacturing',
            7:'Transportation & Utility', 8:'Public Facilities & Institutions', 9:'Open Space & Outdoor Recreation',
            10:'Parking Facilities', 11:'Vacant Land'}

both.loc[both['LandUse'].isin(verbose.keys()), 'LandUse'] = both['LandUse'].map(verbose)
both
```

#### manhattan/brooklyn building category ratio

```python
both['ratio'] = (both.mn/both.bk)
both
```

```python
def mn_greater(ratio):
    if ratio > 1.0:
        return True
    else:
        return False

both['mn_greater'] = both['ratio'].apply(mn_greater)
both
```

#### findings
Manhattan has a greater number of buildings in only a single category: multi-family buildings with an elevator. Whereas Manhattan has a lot of residential units in skyscrapers, necessitating elevators to access them, Brooklyn's multi-family residences tend not to have so many stories.
